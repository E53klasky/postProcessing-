```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Prepare the Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
U = -1 - X**2 + Y
V = 1 + X - Y**2

# 2. Specify the Starting Point
start_point = np.array([[2, 2]])

# 3. Plot the Streamline
fig, ax = plt.subplots()
#ax.streamplot(X, Y, U, V, start_points=start_point)
ax.streamplot(X,Y,U,V)
ax.set_aspect('equal')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# example
# 1. Prepare the Data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
U = -1 - X**2 + Y
V = 1 + X - Y**2

# 2. Specify the Starting Point
start_point = np.array([[2, 2]])

# 3. Plot the Streamline
fig, ax = plt.subplots()
ax.streamplot(X, Y, U, V, start_points=start_point)
#x.streamplot(X,Y,U,V)
ax.set_aspect('equal')
plt.show()
```
